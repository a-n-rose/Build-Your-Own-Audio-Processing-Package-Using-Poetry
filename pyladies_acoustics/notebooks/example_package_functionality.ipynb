I incorporated `scipy.io.wavfile.read` into my example package as well as code from a "programming challenge" (with a slight name adjustment: `adjust_volume` to `adjust_amplitude`).

```python
import pyladies_acoustics as pa
import matplotlib.pyplot as plt
```

```python
pa.__version__
```

#### We can see the basic [`docstring`](https://www.python.org/dev/peps/pep-0257/) that I have written for these functions:

```python
print(pa.audio_fun.load_audio.__doc__)
```

```python
samples_speech, sr_speech = pa.audio_fun.load_audio("../../audio/python.wav")
```

```python
print(pa.audio_fun.adjust_amplitude.__doc__)
```

```python
samples_speech_reduced = pa.audio_fun.adjust_amplitude(
    samples_speech,
    amount=0.5
)
```

### Let's use matplotlib to see that the amplitude values have been adjusted as we expect

```python
plt.plot(samples_speech)
plt.title("Original samples")
plt.show()
```

```python
plt.plot(samples_speech_reduced)
plt.title("Reduced samples")
plt.show()
```

Great! It shows that the amplitude has in face been reduced as we expected.
